```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 25, 10
```

```python
df = pd.read_csv("Macroeconomic_Aggregates.csv", thousands=",")
df.set_index('Year', inplace=True)
df.sort_index(ascending=True, inplace=True)

display(df)
```

```python
df_final = df.diff()[1:]
df_final
```

```python
sns.barplot(x= df_final.index, y=df_final['Net taxes on Products'], color="darkblue")
plt.suptitle('Change in net taxes on products (at constant prices) \nBase year: 2011-12', fontsize=28, fontweight='bold')
plt.xlabel("Years", fontsize=14, fontweight='bold')
plt.ylabel("Change (Rupees Crore)", fontsize=14, fontweight='bold')
plt.show()
```
