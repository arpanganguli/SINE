```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 25, 10
```

```python
df = pd.read_csv("Macroeconomic_Aggregates.csv", thousands=",")
```

```python
df.head()
```

```python
df.set_index('Year', inplace=True)
df.head()
```

```python
df.sort_index(ascending=True, inplace=True)
df
```

```python
df.info()
```

```python
df_final = df.diff()[1:]
df_final
```

```python
sns.barplot(x= df_final.index, y=df_final['Net taxes on Products'], color="darkblue")
plt.suptitle('Change in net taxes on products', fontsize=28, fontweight='bold')
plt.xlabel("Year", fontsize=14, fontweight='bold')
plt.ylabel("Change", fontsize=14, fontweight='bold')
plt.show()
```
